```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
#Quelle: https://learnopencv.com/edge-detection-using-opencv/ 
```

```python
#Grauwerte ber 8bit aufspreizen, um grere Dynamik zu erreichen
#Zur Kontrastverstrkung: viele Werte der Grauwertskala nicht vertreten (siehe Histogramm des originalen Bildes  an den Rndern keine Grauwerte vertreten).
#Grauwerte werden auseinandergezogen, um Abstand zwischen hellstem und dunkelstem Grauwert zu vergrern \autocite{Histo}
Bild = cv2.imread('beispiel_intensity_image.bmp')
iBild = cv2.cvtColor(Bild, cv2.COLOR_RGB2BGR)
Bildequal = iBild.copy()

for i in range(3):
    Bildequal[:,:,i] = cv2.equalizeHist(iBild[:,:,i])      # Bilddynamik ausnutzen und Grauwerte auf 8bit verteilen
plt.imshow(Bildequal)
plt.axis('off')
plt.show()
# Beispielhafter Vergleich fr den blauen Kanal
plt.figure(figsize = (10,4))  
plt.hist(iBild[:,:,2].ravel(), bins=150, color='k', label='Orginalbild')
plt.hist(Bildequal[:,:,2].ravel(), bins=150, color='b', label='Bild nach der Histogrammanpassung')
plt.title('Histogramm')
plt.legend()
plt.show()

cv2.imwrite('equal.png', Bildequal)
```

```python
#2D-Faltung
#2D-Matrix, die das Bild darstellt, mit kleinerer Matrix, dem sogenannten 2D-Kernel falten\autocite{2D}
#Bild wird verwaschen um Kanten besser erkennen zu knnen

#in 50 mikrometer Tiefe
img = cv2.imread('beispiel_intensity_image.bmp')
kernel = np.ones((5,5),np.float32)/25
dst = cv2.filter2D(img,-1,kernel)

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.show()
cv2.imwrite('averaging.png', dst)
```

```python
#in 200 mikrometer Tiefe
img = cv2.imread('beispiel200_intensity_image.bmp')
jBild = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
kernel = np.ones((5,5),np.float32)/25
dst = cv2.filter2D(img,-1,kernel)
Bildequal = jBild.copy()
for i in range(3):
    Bildequal[:,:,i] = cv2.equalizeHist(jBild[:,:,i])

plt.subplot(131),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(132),plt.imshow(dst),plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.subplot(133),plt.imshow(Bildequal),plt.title('Equal')
plt.xticks([]), plt.yticks([])
plt.show()
cv2.imwrite('averaging200.png', dst)
cv2.imwrite('equal200.png', Bildequal)
```

```python
#Anwendung eines Canny Edge Filters
#in 50 mikrometer Tiefe
#gewhlter threshold fr alle Bilder
threshold1 = 1
threshold2 = 250
path = 'beispiel_intensity_image.bmp'
path2 = 'averaging.png'
path3 = 'equal.png'
img = cv2.imread(path,-1)
img2 = cv2.imread(path2,-1)
img3 = cv2.imread(path3,-1)
b,g,r = cv2.split(img)
b2,g2,r2 = cv2.split(img2)
b3,g3,r3 = cv2.split(img3)
edgeImage = cv2.Canny(b,threshold1,threshold2)
edgeImage2 = cv2.Canny(b2,threshold1,threshold2)
edgeImage3 = cv2.Canny(b3,threshold1,threshold2)
plt.subplot(321),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(322),plt.imshow(edgeImage),plt.title('Edges')
plt.xticks([]), plt.yticks([])
plt.subplot(323),plt.imshow(img2),plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.subplot(324),plt.imshow(edgeImage2),plt.title('Edges Averaging')
plt.xticks([]), plt.yticks([])
plt.subplot(325),plt.imshow(img3),plt.title('Equal')
plt.xticks([]), plt.yticks([])
plt.subplot(326),plt.imshow(edgeImage3),plt.title('Edges Equal')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#HPF filters (help in finding edges in images)
#in 50 mikrometer Tiefe
#variation der Interpolation
img = cv2.imread('beispiel_intensity_image.bmp')
img2 = cv2.resize(img, (128, 128), 
                 interpolation=cv2.INTER_CUBIC)
# subtract the original image with the blurred image
# after subtracting add 127 to the total result
hpf2 = img2 - cv2.GaussianBlur(img, (21, 21), 3)+127
img3 = cv2.resize(img, (128, 128), 
                 interpolation=cv2.INTER_BITS)
# subtract the original image with the blurred image
# after subtracting add 127 to the total result
hpf3 = img3 - cv2.GaussianBlur(img, (21, 21), 3)+127
img4 = cv2.resize(img, (128, 128), 
                 interpolation=cv2.INTER_LINEAR)
# subtract the original image with the blurred image
# after subtracting add 127 to the total result
hpf4 = img4 - cv2.GaussianBlur(img, (21, 21), 3)+127
plt.subplot(141),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(142),plt.imshow(hpf2),plt.title('HPF Cubic')
plt.xticks([]), plt.yticks([])
plt.subplot(143),plt.imshow(hpf3),plt.title('HPF Bits')
plt.xticks([]), plt.yticks([])
plt.subplot(144),plt.imshow(hpf4),plt.title('HPF linear')
plt.xticks([]), plt.yticks([])
plt.show()
cv2.imwrite('HPF2.bmp', hpf2)
cv2.imwrite('HPF3.bmp', hpf3)
cv2.imwrite('HPF4.bmp', hpf4)
```

```python
#Anwendung von Sobel und Laplace
#in 50 mikrometer Tiefe
img = cv2.imread('beispiel_intensity_image.bmp')
# resize image
img = cv2.resize(img, (128, 128), interpolation=cv2.INTER_CUBIC)
# convert image to gray scale image
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# apply laplacian blur
laplacian = cv2.Laplacian(gray, cv2.CV_64F)
# sobel x filter where dx=1 and dy=0
sobelx = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=7)
# sobel y filter where dx=0 and dy=1
sobely = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=7)
# combine sobel x and y
sobel = cv2.bitwise_and(sobelx, sobely)
  
# plot images
plt.subplot(2, 2, 1)
plt.imshow(laplacian, cmap='gray')
plt.title('Laplacian')
plt.axis('off')
  
plt.subplot(2, 2, 2)
plt.imshow(sobelx, cmap='gray')
plt.title('SobelX')
plt.axis('off')
  
plt.subplot(2, 2, 3)
plt.imshow(sobely, cmap='gray')
plt.title('SobelY')
plt.axis('off')
  
plt.subplot(2, 2, 4)
plt.imshow(sobel, cmap='gray')
plt.title('Sobel')
plt.axis('off')

plt.show()

cv2.imwrite('Laplacian.bmp', laplacian)
```

```python
#Edge Detection Using OpenCV
#in 50 mikrometer Tiefe
# Read the original image
img = cv2.imread('beispiel_intensity_image.bmp') 
 
# Convert to graycsale
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# Blur the image for better edge detection
img_blur = cv2.GaussianBlur(img_gray, (3,3), 0) 
 
# Sobel Edge Detection
sobelx = cv2.Sobel(src=img_blur, ddepth=cv2.CV_64F, dx=1, dy=0, ksize=5) # Sobel Edge Detection on the X axis
sobely = cv2.Sobel(src=img_blur, ddepth=cv2.CV_64F, dx=0, dy=1, ksize=5) # Sobel Edge Detection on the Y axis
sobelxy = cv2.Sobel(src=img_blur, ddepth=cv2.CV_64F, dx=1, dy=1, ksize=5) # Combined X and Y Sobel Edge Detection
 
# Canny Edge Detection
edges = cv2.Canny(image=img_blur, threshold1=100, threshold2=200) # Canny Edge Detection

# plot images
plt.subplot(2, 3, 1)
plt.imshow(img_gray, cmap='gray')
plt.title('img_gray')
plt.axis('off')

plt.subplot(2, 3, 2)
plt.imshow(img_blur, cmap='gray')
plt.title('img_blur')
plt.axis('off')

# Display Sobel Edge Detection Images
plt.subplot(2, 3, 3)
plt.imshow(sobelx, cmap='gray')
plt.title('SobelX')
plt.axis('off')
  
plt.subplot(2, 3, 4)
plt.imshow(sobely, cmap='gray')
plt.title('SobelY')
plt.axis('off')
  
plt.subplot(2, 3, 5)
plt.imshow(sobel, cmap='gray')
plt.title('Sobel')
plt.axis('off')

# Display Canny Edge Detection Image
plt.subplot(2, 3, 6)
plt.imshow(edges, cmap='gray')
plt.title('Canny Edge Detection')
plt.axis('off')
plt.show()
```
